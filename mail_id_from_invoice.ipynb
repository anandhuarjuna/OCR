```python
import cv2
import numpy as np
import pytesseract
img=cv2.imread(r'D:\Computer Vision\OCR\mailid invoice.jpg')
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow('Invoice',gray_img)
cv2.waitKey()
cv2.destroyAllWindows
```

```python
(thresh,binary_img)=cv2.threshold(gray_img,250,255,cv2.THRESH_BINARY| cv2.THRESH_OTSU)
cv2.imshow('Binary_image',binary_img)
cv2.waitKey()
cv2.destroyAllWindows
```

```python
pytesseract.pytesseract.tesseract_cmd=r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
data=pytesseract.image_to_data(gray_img,output_type=pytesseract.Output.DICT)
print(data)
```

```python
import re
n_boxes=len(data['text'])
for i in range(n_boxes):
    if data['conf'][i]>=40:
        if re.search(r'[\w\.-]+@[\w\.-]+',data['text'][i]):
            x=data['left'][i]
            y=data['top'][i]
            w=data['width'][i]
            h=data['height'][i]
            print(x,y,w,h)  
            img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)   
cv2.imshow('image',img)
cv2.waitKey()
cv2.destroyAllWindows
```
