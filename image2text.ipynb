```python
data=pytesseract.image_to_data(image2,output_type=pytesseract.Output.DICT)
data
```

```python
#to print the boxes in each words
n_boxes=len(data['text'])
for i in range(n_boxes):
    if data['conf'][i]>50:
        x,y,w,h=data['left'][i],data['top'][i],data['width'][i],data['height'][i]
        print(x,y,w,h)
        img=cv2.rectangle(image2,(x,y),(x+w,y+h),(255,0,0),3)
cv2.imshow('Image',image2)
cv2.waitKey()
cv2.destroyAllWindows
```

```python
img=cv2.imread(r'D:\Computer Vision\OCR\test.png')
```

```python
target_word='dog'
data=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
```

```python
#For taking the index of specified word
l=data['text']
word_occ=[i for i, word in enumerate(l) if word.lower()==target_word]
print(word_occ)
```
